## Data Visualization using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=df['Occupation']
y=df['Stress Level']
plt.barh(x,y,color="orange")                 #width gives error in barh function
plt.title('Bar Graph')
plt.xlabel('Stress Level')
plt.ylabel('Occupation')
plt.grid(axis='x') 
plt.show
```

```python
a=df.groupby('BMI Category').count()
```

```python
a
```

```python
lbl=['Overweight', 'Normal', 'Obese', 'Normal Weight']
ex=[0,0.1,0.2,0]
clrs=['lavender','pink','yellow','blue']
plt.pie(a['ID'],labels=lbl,colors=clrs,startangle=120,explode=ex)
plt.title('Pie Chart')
plt.legend()        
plt.show()
```

```python
a=df['Daily Steps']
plt.hist(a,bins=[2000,4000,5000,5500,7000,8000,8500,9000,10000],ec='yellow',color='green')
plt.title('Histogram')
plt.xlabel('Daily Steps')
plt.ylabel('No. of People')
plt.show()
```

## Data Visualization using seaborn

```python
import seaborn as sns
```

```python
sns.countplot(x='Gender',data=df,hue='Sleep Disorder',palette='Set1')
plt.title('Countplot')
```

```python
sns.kdeplot(x="Heart Rate",data=df,color='purple',shade=True)
plt.title('KDE plot')
```

```python
sns.displot(x='Physical Activity Level',bins=6,data=df,kde=True,hue=df['Sleep Disorder'],palette='Set3')
plt.title('Displot')
```

```python
sns.scatterplot(x='Physical Activity Level',y='Daily Steps',data=df,hue='BMI Category')
plt.title('Scatter plot')
```

```python
cor=df.corr(numeric_only=True)
sns.heatmap(cor,annot=True)
plt.title('Correlation Heatmap')
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(data=df,x='Physical Activity Level',y='Sleep Duration',kind='reg')
```
