# STUDENT REGISTRATION SYSTEM

```python
class Std:
    def __init__(self, name, roll, marks1, marks2):
        self.name=name
        self.roll=roll
        self.marks1=marks1
        self.marks2=marks2
    def display(self):
        print("\nname:",self.name)  
        print("roll no:",self.roll) 
        print("marks1:",self.marks1) 
        print("marks2",self.marks2)
       
       
def search(Stds, search_roll):
    x = [i for student in Stds if student.roll == search_roll]
    if x:
        print("\nDetails of searched student:")
        for student in x:
            student.display()
    else:
        print("\nno students found")  
        
def update(Stds, search_roll):
    for student in Stds:
        if student.roll == search_roll:
            name = input("Enter updated name: ")
            mark1 = int(input("Enter updated marks1: "))
            mark2 = int(input("Enter updated marks2: "))
            student.name = name
            student.marks1 = mark1
            student.marks2 = mark2
            print("Student info updated successfully!")
            return
    print("\nno students found")

def delete(Stds,search_roll):
    for student in Stds:
        if student.roll == search_roll:
            Stds.remove(student)
            print("\nstudent details with",search_roll,"roll no. are deleted")
            return
    print("\nno students found")

    
Stds=[]


print("Student Registration System")
while True:
    print("\nOPtions: 1. Accept  2. Display  3. Search  4. Update  5. Delete  6. Exit")
    
    ch=int(input("Enter your choice in numbers:"))     
    
    if ch==1:
        n=str(input("enter name:"))
        r=int(input("enter roll no:"))
        m1=int(input("enter marks1:"))
        m2=int(input("enter marks2:"))
        student=Std(n,r,m1,m2)
        Stds.append(student)
        print("Student Regitered\n")
        
    elif ch==2:
        print("\nStudent list:\n")
        if (len(Stds)==0):
            print("\nno student registered\n")
        for student in Stds:
            student.display()
            
    elif ch==3:
        if (len(Stds)==0):
            print("\nno student found")
        else:
            search_roll=int(input("\nEnter student roll no:"))
            search(Stds,search_roll)
   
    elif ch==4:
        if (len(Stds)==0):
            print("\nno student found")
        else:
            search_roll=int(input("\nEnter student roll no:"))
            update(Stds,search_roll)
               
    elif ch==5:
        if (len(Stds)==0):
            print("\nno student found") 
        else:
            search_roll=int(input("\nEnter student roll no:"))
            delete(Stds,search_roll)
          
            
    elif ch==6:
        break
        
    else:
        print("\nWorng choice. Enter your choice again.")
        
        
            
```
