### Data splitting

```python
x = df.iloc[:,:-1]    #independent
y = df.iloc[:,-1]     #dependent
```

```python
x
```

```python
y
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=2) 
```

```python
x_test
```

```python
y_test
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Data scalling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

# step4) Model Creation: Apply algorithms
# step5) Evaluation & Validtion: Check accuracy

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
li_reg=LinearRegression()                    
li_reg.fit(x_train,y_train)
```

```python
li_reg.score(x_train,y_train)
```

```python
y_pred=li_reg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)                  #accuracy
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
log_reg.score(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier()
rf_class.fit(x_train,y_train)
```

```python
rf_class.score(x_train,y_train)
```

```python
y_pred = rf_class.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_model  = GaussianNB().fit(x_train,y_train)
nb_model.score(x_train,y_train)
```

```python
y_pred = nb_model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

### SVC

```python
from sklearn.svm import SVC
svc_model = SVC().fit(x_train,y_train)
svc_model.score(x_train,y_train)
```

```python
y_pred = svc_model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

### Conclusion: Logistic Regression has the highest accuracy (0.85) among all the above applied algorithms.
